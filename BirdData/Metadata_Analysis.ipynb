```python
def draw_bodymass_scatter():
    CommunityDivision_Info_Weighted = pd.read_csv("Rankularity_output/CommunityDivision_Info_Weighted.csv")
    community = CommunityDivision_Info_Weighted['0.6']
    Species = CommunityDivision_Info_Weighted['Species']
#     print("community = ", list(community), len(community))
    metadata = pd.read_csv("BirdSpecies_metadata.csv")
    species = list(metadata['species'])
    log_mass = list(metadata['log.mass'])
    BT_rank = list(metadata['bradley.terry.rank'])
    print(max(BT_rank))
    
    species_mass_dict = {}
    for i in range(len(species)):
        species_mass_dict[species[i]] = (log_mass[i], BT_rank[i])
        
    body_mass_metadata = []
    BT_metadata = []
    for i in range(len(Species)):
        body_mass_metadata.append(species_mass_dict[Species[i]][0])
        BT_metadata.append(species_mass_dict[Species[i]][1])
        
    #plt.plot(list(community-0.05), np.array(body_mass_metadata),'go', markersize = 10, alpha=0.3)
    plt.plot(list(community+0.05), list(BT_metadata),'ro', markersize = 10,label="Loss",alpha=0.2)
    plt.xticks([1,2,3,4],[1,2,3,4], fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.xlabel("Community", fontsize = 15)
    #plt.ylabel("Log(body mass)", fontsize = 15)
    plt.ylabel("Bradley Terry Rank", fontsize = 15)
    #plt.legend(fontsize = 12,loc="upper left")
    #plt.ylim(1,16)
    #,4,5,6,7,8,9,10]
    plt.show()
```

```python
draw_bodymass_scatter()
```
