```python
def create_network():
    bird_data = pd.read_csv("SpeciesInteraction.csv")
    source = list(bird_data['source'])
    target = list(bird_data['target'])
    interactions = defaultdict(int)
    
    for i in range(len(source)):
        interactions[(source[i],target[i])] += 1
    
    G = nx.DiGraph()
    for eachpair in interactions:
        G.add_edge(eachpair[0],eachpair[1],weight = interactions[eachpair])
     
    print("Number of nodes = ", G.number_of_nodes())
    print("Number of edges = ", G.number_of_edges())
    return G
    #print(G.edges(data=True))
    
```

```python
def save_Rankularity_output(): # This is specifically for 2013.
    G = create_network()
    graph_and_hashes,node_division,alphawise_info, R = Weighted_Rankularity_Sankey.get_Rankularity_results(G, 4)
    graph_path = "Rankularity_output/graph_and_hashes.pkl"
    pickle_out = open(graph_path,"wb")
    pickle.dump(graph_and_hashes, pickle_out)
    pickle_out.close()
    division_path = "Rankularity_output/node_division.pkl"
    pickle_out = open(division_path,"wb")
    pickle.dump(node_division, pickle_out)
    pickle_out.close() 
    info_path = "Rankularity_output/alphawise_info.pkl"
    pickle_out = open(info_path,"wb")
    pickle.dump(alphawise_info, pickle_out)
    pickle_out.close()    
```

```python
# Using MVR
save_Rankularity_output()
```
